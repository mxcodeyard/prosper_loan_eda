## 2.0 Introduction

There are 81 features with over 100,000 rows. We will not be going through all of these. Instead, we will be going through the key variables and selecting those that might seem relevant to the potential questions we seek to answer.

Before going further, we will first try to gain a better understanding of what this dataset is about.

### About the Dataset
In 2005, Prosper was started in the United States as the first <b>peer-to-peer lending marketplace</b> backed by leading investors including Sequoia Capital, Francisco Partners, Institutional Venture Partners, and Credit Suisse NEXT Fund. Fixed-rate and fixed-term loan from USD2,000 to USD40,000 are being offered to borrowers while lenders can invest in loans and earn attractive returns while Prosper takes care of the loan services on their behalf respectively.

The dataset contains information regarding borrowers and lenders who used the Prosper loan services from 2005 to 2014.

>#### Reference(s):
https://www.prosper.com/about

### 2.1 Features of Interest


<br><b>ListingNumber</b> - The number that uniquely identifies the listing to the public as displayed on the website.
<br>
<br><b>ListingCreationDate</b> - The date the listing was created.
<br>
<br><b>Term</b> - Length of the loan expressed in months.
<br>

<br><b>LenderYield</b> - The Lender yield on the loan. Lender yield is equal to the interest rate on the loan less the servicing fee.
<br>
<br><b>ListingCategory</b> - The category of the listing that the borrower selected when posting their listing: 0 - Not Available, 1 - Debt Consolidation, 2 - Home Improvement, 3 - Business, 4 - Personal Loan, 5 - Student Use, 6 - Auto, 7- Other, 8 - Baby&Adoption, 9 - Boat, 10 - Cosmetic Procedure, 11 - Engagement Ring, 12 - Green Loans, 13 - Household Expenses, 14 - Large Purchases, 15 - Medical/Dental, 16 - Motorcycle, 17 - RV, 18 - Taxes, 19 - Vacation, 20 - Wedding Loans

<br>
<br><b>EmploymentStatus</b> - The employment status of the borrower at the time they posted the listing.
<br>
<br><b>EmploymentStatusDuration</b> - The length in months of the employment status at the time the listing was created.
<br>
<br><b>IsBorrowerHomeowner</b> - A Borrower will be classified as a homowner if they have a mortgage on their credit profile or provide documentation confirming they are a homeowner.
<br>
<br><b>CreditScoreRangeLower</b> - The lower value representing the range of the borrower's credit score as provided by a consumer credit rating agency.
<br>
<br><b>CreditScoreRangeUpper</b> - The upper value representing the range of the borrower's credit score as provided by a consumer credit rating agency. 
<br>
<br><b>FirstRecordedCreditLine</b> - The date the first credit line was opened.
<br>
<br><b>TotalInquiries</b> - Total number of inquiries at the time the credit profile was pulled.
<br>
<br><b>CurrentDelinquencies</b> - Number of accounts delinquent at the time the credit profile was pulled.
<br>
<br><b>AmountDelinquent</b>	- Dollars delinquent at the time the credit profile was pulled.
<br>
<br><b>BankcardUtilization</b>	- The percentage of available revolving credit that is utilized at the time the credit profile was pulled.

<br><b>StatedMonthlyIncome</b>	- The monthly income the borrower stated at the time the listing was created.

### _Matrices Applicable Only Before July 2009_

### _Matrices Implemented After July 2009_

<br><b>EstimatedEffectiveYield</b> - Effective yield is equal to the borrower interest rate (i) minus the servicing fee rate, (ii) minus estimated uncollected interest on charge-offs, (iii) plus estimated collected late fees.  Applicable for loans originated after July 2009.
<br>
<br><b>EstimatedLoss</b> - Estimated loss is the estimated principal loss on charge-offs. Applicable for loans originated after July 2009.
<br>
<br><b>EstimatedReturn</b> - The estimated return assigned to the listing at the time it was created. Estimated return is the difference between the Estimated Effective Yield and the Estimated Loss Rate. Applicable for loans originated after July 2009.
<br>

<br>
<br><b>ProsperRating (Alpha)</b> - The Prosper Rating assigned at the time the listing was created between AA - HR.  Applicable for loans originated after July 2009.
<br>
<br><b>ProsperScore</b> - A custom risk score built using historical Prosper data. The score ranges from 1-10, with 10 being the best, or lowest risk score.  Applicable for loans originated after July 2009.

**Credit Score** - The value representing the range of the borrower's credit score as provided by a consumer credit rating agency.

There are two features related to credit score, lower range and upper range, which can be converted to only one credit score feature by taking the average. A new feature called **AvgCreditScore** will be created in place of those two credit score variables.

```python
# no. of missing values for CreditScoreRangeLower & CreditScoreRangeUpper
df['CreditScoreRangeLower'].isnull().sum(),df['CreditScoreRangeUpper'].isnull().sum()
```

### 3.7 Loan Status

```python
# loan status category frequency
df['LoanStatus'].value_counts()
```

There are 6 types of <b>Past Due</b> loans with different timeframes. To simplify the analysis process, all past due loans will be merged into one type except when it is past 120 days. Loans past 120 days are considered charged off and hence, will be recategorized into chargedoff. 

Considering the amount of <b>Cancelled</b> loans are negligible relative to others and did not move to the next stage of process application, these loans will be put together with completed loans.

Moreoever, since <b>FinalPaymentInProgress</b> loans do not amount to the majority of the loans and are about to be completed but nevertheless still active, these loans will be placed into current loans.

><br>Reference(s):
<br>https://prosper.zendesk.com/hc/en-us/articles/208500546-What-is-a-charge-off-

```python
# recategorizing/replacing loan status
loans_recat = {
             'Past Due (1-15 days)' : 'Past Due' ,
             'Past Due (16-30 days)' : 'Past Due',
             'Past Due (31-60 days)' : 'Past Due',
             'Past Due (61-90 days)' : 'Past Due',
             'Past Due (91-120 days)' : 'Past Due',
             'Past Due (>120 days)' : 'Chargedoff',
             'FinalPaymentInProgress' : 'Current',
             'Cancelled' : 'Completed'
            }

df.replace({'LoanStatus':loans_recat},inplace=True)
```

```python
# set loan status categorical order
cat = ['Completed','Current','Past Due','Chargedoff','Defaulted']
cat_order = pd.api.types.CategoricalDtype(categories=cat, ordered=True)
df['LoanStatus'] = df['LoanStatus'].astype(cat_order)

# loan status count after reorganizing
loan_count = df['LoanStatus'].value_counts().sort_index()
loan_count
```

```python
# loan status plot
total_count = df.shape[0]
max_loan_prop = (loan_count/total_count).max()

loan_yticks = np.arange(0,max_loan_prop+0.1,0.1)
loan_ylabel = ['{:.2f}'.format(x) for x in loan_yticks]

sb.countplot(df['LoanStatus'],color=base_color)

plt.xlabel('Loan Status')
plt.yticks( loan_yticks*total_count ,loan_ylabel)
plt.ylabel('Proportion')
plt.grid(linestyle='dotted');

for i in range(loan_count.shape[0]):

    count = loan_count[i]
    pct_string = '{:0.1f}%'.format( (count/total_count) * 100 )
    plt.text(i,count+250, pct_string,ha='center')
```

## 4.1 BorrowerAPR, BorrowerRate & LenderYield

The heatmap shows that **BorrowerAPR, BorrowerRate & LenderYield** have almost identical correlation with other features. Hence, they should have similar graphs. So, what's the difference between these three features?

**BorrowerAPR** - The Borrower's Annual Percentage Rate (APR) for the loan.\
**BorrowerRate** - The Borrower's interest rate for this loan.\
**LenderYield** - The Lender yield on the loan. Lender yield is equal to the interest rate on the loan less the servicing fee.

```python
# missing values if any for BorrowerAPR, BorrowerRate & LenderYield
df['BorrowerAPR'].isnull().sum(), df['BorrowerRate'].isnull().sum(), df['LenderYield'].isnull().sum()
```

In **Section 3.1 - Borrower APR**, it was mentioned that the differences between Borrower Rate and Borrower APR is that Borrower APR includes other fees on top of the interest rate associated with the loan. _i.e. Prosper Loan taking a certain percentage as a charging fee from the loan proceeds._ Also, the distribution plot shows that Borrower APR is higher than both Borrower Rate and Lender Yield which confirms our initial hypothesis. Moreover, Lender Yield lags slightly behind Borrower Rate as it is interest rate minus servicing fee.

With **Borrower APR having the higher rate** compared to the other two, Borrower APR will be used instead for the remaining of the analysis while both Borrower Rate and Lender Yield will be dropped.

### 4.2 BorrowerAPR vs. ProsperRating & ProsperScore

**Prosper Rating** & **Prosper Score** are two similar features describing the ratings of the loan **after July 2019**. Heatmap shows that both are strongly correlated with Borrower APR. The differences between the two variables are as follow: 

>**Prosper Rating** - Rating assigned at the time the listing was created: **0 - N/A, 1 - HR, 2 - E, 3 - D, 4 - C, 5 - B, 6 - A, 7 - AA**. It is based on Prosper's proprietary rating system. The ratings are categorized by different thresholds of the **estimated avg. annual loss rate** which is determined by two metrics:
>>(i) **Borrower's credit score**, obtained from an official credit reporting agency, and\
(ii) **Prosper Score**, figured in-house based on the Prosper population.

>**Prosper Score** - A custom risk score built using historical Prosper data. The **score ranges from 1-10, with 10 being the best, or lowest risk score**. Key metrics used to determine the prosper score are:
>>(i) **Number of trades**\
(ii) **Number of delinquent accounts**\
(iii) **Number of inquiries**\
(iv) **Number of recently opened trades**\
(v) **Amount of available credit on bankcards**\
(vi) **Bankcard utilization**

>Reference(s):\
https://www.prosper.com/invest/how-to-invest/prosper-ratings/?mod=article_inline \
https://www.prosper.com/help/topics/general-prosper_score.aspx


```python
# prosper rating dictionary 
pr_dict = {
           1:'HR',
           2:'E',
           3:'D',
           4:'C',
           5:'B',
           6:'A',
           7:'AA'
          }
```

```python
# color palette for plot
color_pal = sb.color_palette()[2]


fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

sb.catplot(data=df,x='ProsperRating (numeric)',y='BorrowerAPR',kind='box',color=color_pal,ax=axes[0])
axes[0].set_xlabel('Prosper Rating\n(High to Low Risk)')
axes[0].set_ylabel("Borrower's APR")
axes[0].set_title('Prosper Rating (Loans After July 2019)')

plt.setp(axes[0], xticks=[y for y in range(len(pr_dict))],
         xticklabels=pr_dict.values())

sb.catplot(data=df,x='ProsperScore',y='BorrowerAPR',kind='box',color=color_pal,ax=axes[1])
axes[1].set_xlabel('Prosper Score\n(High to Low Risk)')
axes[1].set_ylabel("Borrower's APR")
axes[1].set_title('Prosper Score (Loans After July 2019)')
plt.close(2)
plt.close(3);
```


For loans after July 2009, the plot above shows that loans with lower prosper score or riskier loans cost more than those considered low risk. Because Prosper Rating shows a perfect correlation to Borrower's APR, it seems that Prosper uses Prosper Rating to determine the Borrower's APR for the loans.
